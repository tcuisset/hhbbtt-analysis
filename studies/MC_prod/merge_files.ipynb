```python
import ROOT
import JupyROOT.helpers.utils
JupyROOT.helpers.utils.GetGeometryDrawer = lambda:None # magic fix to make ROOT faster
from pathlib import Path
import uuid
```

```python
tmp_datasets_base = "/data_CMS/cms/cuisset/ZHbbtautau/framework/nanoaod_base_analysis/data/tmp/ul_2018_ZZ_v12/"
preprocess_output_base = lambda dataset, index : f"/grid_mnt/data__data.polcms/cms/cuisset/cmt/PreprocessRDF/ul_2018_ZZ_v12/{dataset}/cat_base_selection/prod_240728/data_{index}.root"
datasets = "data_etau_a, data_etau_b, data_etau_c, data_etau_d, data_mutau_a, data_mutau_b, data_mutau_c, data_mutau_d, data_tau_a, data_tau_b, data_tau_c, data_tau_d, dy, dy_0j, dy_1j, dy_2j, dy_ptz1, dy_ptz2, dy_ptz3, dy_ptz4, dy_ptz5, dy_ptz6, ewk_wminus, ewk_wplus, ewk_z, ggf_sm, ggH_ZZ, st_antitop, st_top, st_tw_antitop, st_tw_top, tt_dl, tt_fh, tth_tautau, tt_sl, ttwh, ttw_lnu, ttw_qq, ttww, ttwz, ttzh, ttz_llnunu, ttz_qq, ttzz, wjets_ht1, wjets_ht2, wjets_ht3, wjets_ht4, wjets_ht5, wjets_ht6, wjets_ht7, wjets_ht8, wminush_htt, wplush_htt, ww_llnunu, ww_lnuqq, ww_qqqq, www, wwz, wz_lllnu, wz_llqq, wz_lnununu, wz_lnuqq, wzz, zh_hbb_zll, zh_htt, zz_bbtt, zz_dl, zz_fh, zz_lnu, zz_qnu, zz_sl_background, zz_sl_signal, zzz".split(", ")
# GluGluToXToZZTo2B2Tau_M1000, GluGluToXToZZTo2B2Tau_M1100, GluGluToXToZZTo2B2Tau_M1200, GluGluToXToZZTo2B2Tau_M1300, GluGluToXToZZTo2B2Tau_M1400, GluGluToXToZZTo2B2Tau_M1500, GluGluToXToZZTo2B2Tau_M1600, GluGluToXToZZTo2B2Tau_M1700, GluGluToXToZZTo2B2Tau_M1800, GluGluToXToZZTo2B2Tau_M1900, GluGluToXToZZTo2B2Tau_M200, GluGluToXToZZTo2B2Tau_M2000, GluGluToXToZZTo2B2Tau_M210, GluGluToXToZZTo2B2Tau_M220, GluGluToXToZZTo2B2Tau_M2200, GluGluToXToZZTo2B2Tau_M230, GluGluToXToZZTo2B2Tau_M240, GluGluToXToZZTo2B2Tau_M2400, GluGluToXToZZTo2B2Tau_M250, GluGluToXToZZTo2B2Tau_M2500, GluGluToXToZZTo2B2Tau_M260, GluGluToXToZZTo2B2Tau_M2600, GluGluToXToZZTo2B2Tau_M270, GluGluToXToZZTo2B2Tau_M280, GluGluToXToZZTo2B2Tau_M2800, GluGluToXToZZTo2B2Tau_M300, GluGluToXToZZTo2B2Tau_M3000, GluGluToXToZZTo2B2Tau_M320, GluGluToXToZZTo2B2Tau_M350, GluGluToXToZZTo2B2Tau_M3500, GluGluToXToZZTo2B2Tau_M360, GluGluToXToZZTo2B2Tau_M400, GluGluToXToZZTo2B2Tau_M4000, GluGluToXToZZTo2B2Tau_M450, GluGluToXToZZTo2B2Tau_M4500, GluGluToXToZZTo2B2Tau_M500, GluGluToXToZZTo2B2Tau_M5000, GluGluToXToZZTo2B2Tau_M550, GluGluToXToZZTo2B2Tau_M600, GluGluToXToZZTo2B2Tau_M650, GluGluToXToZZTo2B2Tau_M700, GluGluToXToZZTo2B2Tau_M750, GluGluToXToZZTo2B2Tau_M800, GluGluToXToZZTo2B2Tau_M850, GluGluToXToZZTo2B2Tau_M900
def getEventParentFile(dataset, index=0):
    """ returns nb of entries and folder name of the InputData NanoAOD skim """
    with open(tmp_datasets_base + dataset + ".txt", "rt") as f:
        parent_file = f.readlines()[index]
    if len(parent_file) < 5 or parent_file[0] != "/":
        raise RuntimeError()
    if parent_file[-1] == "\n":
        parent_file = parent_file[:-1]
    if parent_file.startswith("/eos/cms"):
        parent_file = "root://eoscms.cern.ch//" + parent_file
    elif parent_file.startswith("/eos/grif"):
        parent_file = "root://eos.grif.fr//" + parent_file
    print(parent_file)
    try:
        f = ROOT.TFile.Open(parent_file)
        return f.Get("Events").GetEntriesFast(), Path(parent_file).parent.name
    finally:
        f.Close()
def getEventsProcessedFile(dataset, index=0):
    try:
        f = ROOT.TFile.Open(preprocess_output_base(dataset, index))
        return f.Get("Events").GetEntriesFast()
    finally:
        f.Close()
def getDatasetSelectionEfficiency(dataset):
    num = getEventsProcessedFile(dataset) 
    den, hleprare_skim_name = getEventParentFile(dataset)
    if num > den: raise ValueError()
    if num < 50: print("LOW STAT")
    return num/den, hleprare_skim_name
```

```python
effs_nba, effs_hleprare = {}, {} # key can be NBA dataset name or hleprare skim dataset name
map_nba_hleprare = {}
for dataset in datasets:
    eff, hleprare_skim_name = getDatasetSelectionEfficiency(dataset)
    effs_nba[dataset] = eff
    effs_hleprare[hleprare_skim_name] = eff
    map_nba_hleprare[dataset] = hleprare_skim_name
effs_nba
```

```python
import json
with open("dataset_selection_efficiency.json", "w") as f:
    json.dump({
        "effs_nba" : effs_nba,
        "effs_hleprare" : effs_hleprare,
        "map_nba_hleprare" : map_nba_hleprare
    }, f)
```

```python
target_nano_sizes = {}
for dataset in datasets:
    # eff is the selection effiency. For most datasets it's extracted from PreprocessRDF, others have hardcoded numbers 
    if dataset.startswith("GluGluToXToZZ"):
        eff = 0.5
    elif dataset in ["dy_ptz6", "dy_ptz5", "dy_ptz4"] + [f"wjets_ht{i}" for i in range(5, 9)]:
        eff = 0.1
    elif dataset in ["dy_ptz3", "dy_ptz2"]+ [f"wjets_ht{i}" for i in range(3, 5)]:
        eff = 0.05
    else:
        eff = effs[dataset]
    if eff < 0.05:
        eff *= 3
    eff = max(eff, 0.01)
    target_nano_sizes[map_nba_hleprare[dataset]] = 100 / eff 
target_nano_sizes # size in MB of input nano
```
